```python
#The Trend of each topic's change along with time is showed below. 
import pandas as pd
import numpy as np
import warnings
documents=pd.read_csv("https://raw.githubusercontent.com/s853348481/MDA/main/MDA_datasets/documents2.csv")
import matplotlib.pyplot as plt

df2 = documents.groupby(['topic','year']).size().unstack(fill_value=0).stack().to_frame().reset_index()
df2.columns = ['topic','year','count']
count1 = list(df2[df2['topic']==0]['count'])
count2 = list(df2[df2['topic']==1]['count'])
count3 = list(df2[df2['topic']==2]['count'])
count4 = list(df2[df2['topic']==3]['count'])
count5 = list(df2[df2['topic']==4]['count'])
count6 = list(df2[df2['topic']==5]['count'])

year = list(df2[df2['topic']==1]['year'])
plt.plot(year,count1,label='topic0:Immigrants')
plt.plot(year,count2,label='topic1:Veteran')
plt.plot(year,count3,label='topic2:Nuclear')
plt.plot(year,count4,label='topic3:Terrorist')
plt.plot(year,count5,label='topic4:Education and jobs')
plt.plot(year,count6,label='topic5:Insurance')
plt.legend()
plt.show()
```

Trend of topics along with time
* Topic0 'Immigrants' has sharp rise starting from 2008. Obama started first presidency in 2008. As a president from democratic party, Obama has been a supporter of American immigrant policies. It also can be seen in the sentiment analysis section, that speeches under topic0 are more 'very positive' rather than 'positive' or 'less positive'. But as a minority president, he may want to be modest during the campaigns to get support from the conservative voters. Then he started to push his new immigrants policy after he became president. This brought more immigrant topics into his speeches. Then topic0 has a second rise from 2014.  

* Topic1 ‘Veteran’ has an overall peak in 2010. It is probably because the American government made a decision to fully withdraw the army from Iraq in 2010. As one of the most important achievements during the presidency, it is natural that Obama focused on this topic during the time period of 2010. 

* Topic2: ‘Nuclear’. The number of speeches relating to this topic are smaller than other topics. It has an overall peak in 2015,  probably because the American government had a major Nuclear deal with Iran in 2015. 

* Topic3:'Terrorist'. This topic has a rapid rise starting from 2014, and reaches its peak in 2015. A further correlation analysis will be conducted below. It can be seen from the sentiment analysis section that speeches under topic3 are more 'less positive' rather than 'positive' or 'very positive'.

* Topic4 ‘Education and jobs’ and Topic5 ‘Insurance’ started to rise from 2008. The number of speeches regarding those topics is roughly evenly distributed from 2010 to 2016.
